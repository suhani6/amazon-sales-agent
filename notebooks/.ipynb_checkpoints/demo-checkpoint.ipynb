```python
# Step 1: Import the required library
import pandas as pd

# Step 2: Load the original raw CSV (update path if needed)
file_path = "../data/amazon_sales.csv"
df = pd.read_csv(file_path)

# Step 3: Normalize column names (lowercase + underscores)
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Step 4: Parse the 'order_date' column into datetime format
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')

# Step 5: Drop rows with critical missing data
df.dropna(subset=['order_date', 'product_name', 'total_sales'], inplace=True)

# Step 6: Fill non-critical missing values (optional)
df['region'] = df['region'].fillna("Unknown")
df['quantity_sold'] = df['quantity_sold'].fillna(0)

# Step 7: Save cleaned data to a new CSV (optional for reuse)
cleaned_path = "../data/clean_amazon_sales.csv"
df.to_csv(cleaned_path, index=False)

# Step 8: Preview cleaned data
df.head()
```
